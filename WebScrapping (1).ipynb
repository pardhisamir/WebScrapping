### get the name of all customers

```python
names=soup.find_all('p',class_='_2sc7ZR _2V5EHH')
names
```

```python
customer_names=[]
for i in range(0,len(names)):
    customer_names.append(names[i].get_text())
```

```python
customer_names
```

### comments

```python
comments=soup.find_all('p',class_='_2-N8zT')
comments
```

```python
comment=[]
for i in range(0,len(comments)):
    comment.append(comments[i].get_text())
```

```python
comment
```

### stars

```python
star=soup.find_all('div',class_='_3LWZlK _1BLPMq')
star
```

```python
stars=[]
for i in range(0,len(star)):
    stars.append(star[i].get_text())
```

```python
stars
```

### reviews

```python
review=soup.find_all('div',class_='t-ZTKy')
review
```

```python
reviews=[]
for i in range(0,len(review)):
    reviews.append(review[i].get_text())
```

```python
reviews
```

### Data Frame

```python
d={'Names':customer_names,
  'comments':comment,
  'stars':stars,
  'Reviews':reviews}
```

```python
d
```

```python
df=pd.DataFrame(d)
```

```python
df
```

```python
df.to_csv('Flipkart Reviews.csv')
```
